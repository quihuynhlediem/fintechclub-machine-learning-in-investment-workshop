```python
import pandas as pd
import numpy as np

# Load the test data and predicted data
test_data = pd.read_csv('test_data.csv')
predicted_data = pd.read_csv('predicted_series_rnn_block10_w50.csv')

# Drop the extra column from the predicted data
predicted_data = predicted_data.drop(columns=['Unnamed: 0'])

# Remove quotation marks from the predicted data columns
predicted_data.columns = predicted_data.columns.str.replace('"', '')

# Print the cleaned column names to confirm
print("Cleaned Test Data Columns:")
print(test_data.columns)

print("Cleaned Predicted Data Columns:")
print(predicted_data.columns)

# Now, calculate distances between actual and predicted values for each stock pair
distances = {}
for stock_pair in test_data.columns:
    actual_values = test_data[stock_pair].values
    predicted_values = predicted_data[stock_pair].values
    
    # Ensure the lengths match by trimming the longer dataset
    min_length = min(len(actual_values), len(predicted_values))
    actual_values = actual_values[:min_length]  # Trim test data
    predicted_values = predicted_values[:min_length]  # Trim predicted data if needed
    
    # Calculate absolute difference (distance) between predicted and actual values
    distance = np.abs(predicted_values - actual_values)
    distances[stock_pair] = distance

# Define the recurrence rates (percentiles)
recurrence_rates = [20, 25, 30, 35, 40, 45]

# Create a DataFrame to store the epsilon thresholds
epsilon_df = pd.DataFrame(columns=['Stock Pair'] + [f'{rate}%' for rate in recurrence_rates])

# For each stock pair, calculate the thresholds based on percentiles
for stock_pair, distance in distances.items():
    thresholds = [np.percentile(distance, rate) for rate in recurrence_rates]
    epsilon_df = epsilon_df.append(pd.Series([stock_pair] + thresholds, index=epsilon_df.columns), ignore_index=True)

# Save the epsilon thresholds to a CSV file
epsilon_df.to_csv('epsilon_test1.csv', index=False)

print("Generated epsilon_test1.csv successfully!")
```
